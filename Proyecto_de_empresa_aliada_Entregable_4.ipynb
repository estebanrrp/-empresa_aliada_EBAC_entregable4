Creando estructura de la base de datos con las siguientes tablas:

DIM_PRODUCT: contiene informacin de productos.

DIM_SEGMENT: contiene atributos de segmentacin.

DIM_CALENDAR: contiene fechas y semanas.

FACT_SALES: contiene los hechos de ventas.

```python
import pandas as pd
import sqlite3

# Cargar los archivos CSV
df_product = pd.read_csv('DIM_PRODUCT.csv', encoding='latin-1')
df_segment = pd.read_csv('DIM_SEGMENT.csv', encoding='latin-1')
df_calendar = pd.read_csv('DIM_CALENDAR.csv', encoding='latin-1')
df_sales = pd.read_csv('FACT_SALES.csv', encoding='latin-1')

# Crear la base de datos SQLite
conn = sqlite3.connect('Ventas.db')

# Exportar cada DataFrame como tabla
df_product.to_sql('DIM_PRODUCT', conn, if_exists='replace', index=False)
df_segment.to_sql('DIM_SEGMENT', conn, if_exists='replace', index=False)
df_calendar.to_sql('DIM_CALENDAR', conn, if_exists='replace', index=False)
df_sales.to_sql('FACT_SALES', conn, if_exists='replace', index=False)

# Confirmar que las tablas se crearon
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("Tablas creadas:", cursor.fetchall())

"""
# Cerrar conexin
conn.close()
"""
```

Observando cuntos registros hay en cada tabla

```python
tablas = ['DIM_PRODUCT', 'DIM_SEGMENT', 'DIM_CALENDAR', 'FACT_SALES']

for tabla in tablas:
    query = f"SELECT COUNT(*) AS total_registros FROM {tabla};"
    resultado = pd.read_sql_query(query, conn)
    print(f"{tabla}:")
    print(resultado)
    print("-" * 40)
```

Mostrar los primeros 5 registros de cada tabla

```python
for tabla in tablas:
    query = f"SELECT * FROM {tabla} LIMIT 5;"
    resultado = pd.read_sql_query(query, conn)
    print(f"Primeros registros de {tabla}:")
    print(resultado)
    print("-" * 40)
```

Ver las columnas y tipos de datos de cada tabla

```python
for tabla in tablas:
    query = f"PRAGMA table_info({tabla});"
    resultado = pd.read_sql_query(query, conn)
    print(f"Estructura de {tabla}:")
    print(resultado)
    print("-" * 40)
```

# Realizando uniones entre tablas

Unir ventas con calendario

```python
df_sales_calendar = pd.merge(df_sales, df_calendar, on='WEEK', how='inner')
```

Unir con productos

```python
df_sales_calendar_product = pd.merge(df_sales_calendar, df_product, left_on='ITEM_CODE', right_on='ITEM', how='inner')
```

Unir con segmentos

```python
df_final = pd.merge(
    df_sales_calendar_product,
    df_segment,
    on=['CATEGORY', 'ATTR1', 'ATTR2', 'ATTR3', 'FORMAT'],
    how='inner'
)
```

Verificando resultado

```python
print(df_final.head())
print(df_final.info())
```

# Generando insights clave sobre las ventas por categora, regin y periodo de tiempo.

Observando ventas por categora

```python
ventas_por_format = df_final.groupby('FORMAT').agg({
    'TOTAL_UNIT_SALES': 'sum',
    'TOTAL_VALUE_SALES': ['sum', 'mean']
}).reset_index()

ventas_por_format.columns = ['FORMAT', 'Total_Unidades', 'Total_Ingresos', 'Ingreso_Promedio']
print("Ventas por FORMAT:")
print(ventas_por_format.sort_values(by='Total_Ingresos', ascending=False))
```

Es posible observar que el formato liquido es el formato con mayores ingresos de todas las presentaciones ofrecidas

Observando ventas por regin

```python
ventas_por_region = df_final.groupby('REGION').agg({
    'TOTAL_UNIT_SALES': 'sum',
    'TOTAL_VALUE_SALES': 'sum'
}).reset_index()

print("Ventas por REGION:")
print(ventas_por_region.sort_values(by='TOTAL_VALUE_SALES', ascending=False))
```

En total de ventas, es posible observar que el rea 2 representa el rea con mayores ventas de todo el pas, representando casi el doble que el rea 4

Ventas por periodo de tiempo (mes y ao)

```python
ventas_por_mes = df_final.groupby(['YEAR', 'MONTH']).agg({
    'TOTAL_UNIT_SALES': 'sum',
    'TOTAL_VALUE_SALES': 'sum'
}).reset_index()

print(ventas_por_mes.sort_values(by=['YEAR', 'MONTH']))
```

Es posible observar que Enero de 2022 represent el mes con ms ventas del priodo analizado. Por otro lado, es posible observar que desde 2022, las ventas fueron cayendo hasta Julio de 2023

Observar una combinacin: categora + mes, es decir, qu categora vendi ms en un determinado mes y ao y observando sus valores en cada ao para comprobar

```python
df_filtered_years = df_final[df_final['YEAR'].isin([2022, 2023])]

ventas_attr1_mes_filtrado = df_filtered_years.groupby(['ATTR1', 'YEAR', 'MONTH']).agg(
    Total_Value_Sales=('TOTAL_VALUE_SALES', 'sum')
).reset_index()

ventas_max_attr1_mes_filtrado = ventas_attr1_mes_filtrado.loc[ventas_attr1_mes_filtrado.groupby('ATTR1')['Total_Value_Sales'].idxmax()]

print("Ventas mximas por ATTR1 en su mes (2022 y 2023):")
display(ventas_max_attr1_mes_filtrado.sort_values(by='Total_Value_Sales', ascending=False))
```

```python
ventas_attr1_ao = df_final.groupby(['ATTR1', 'YEAR']).agg(
    Total_Value_Sales=('TOTAL_VALUE_SALES', 'sum')
).reset_index()

print("Ventas por ATTR1 y Ao:")
display(ventas_attr1_ao.sort_values(by=['ATTR1', 'YEAR']))
```

Se observa a continuacion la clara preferencia por el Cloro en las ventas de ambos aos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=ventas_attr1_ao, x='ATTR1', y='Total_Value_Sales')
plt.xticks(rotation=45)
plt.title('Ingresos por ATTR1')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ventas_por_brand_ao = df_final.groupby(['BRAND', 'YEAR']).agg(
    Total_Value_Sales=('TOTAL_VALUE_SALES', 'sum')
).reset_index()

# Obtener las top 10 marcas en general para mantener consistencia en los aos mostrados
top_brands_overall = df_final.groupby('BRAND')['TOTAL_VALUE_SALES'].sum().sort_values(ascending=False).head(10).index

ventas_por_brand_ao_top = ventas_por_brand_ao[ventas_por_brand_ao['BRAND'].isin(top_brands_overall)]

plt.figure(figsize=(14, 7))
sns.barplot(data=ventas_por_brand_ao_top, x='BRAND', y='Total_Value_Sales', hue='YEAR')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Ingresos por Marca y Ao')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ventas_por_brand = df_final.groupby('BRAND').agg(
    Total_Value_Sales=('TOTAL_VALUE_SALES', 'sum')
).reset_index()

# Seleccionar las top 10 marcas por ingresos
top_10_brands = ventas_por_brand.sort_values(by='Total_Value_Sales', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_brands, x='BRAND', y='Total_Value_Sales')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Ingresos por Marca')
plt.tight_layout()
plt.show()
```

### Insights Clave de los Resultados:

Basndonos en el anlisis de los datos de ventas, podemos destacar los siguientes puntos clave:

*   **Formato de Mayor Ingreso:** El formato "LIQUIDO" es consistentemente el que genera mayores ingresos en comparacin con otros formatos como GEL, POLVO, BARRA y TOALLAS. Esto sugiere una fuerte preferencia del mercado por los productos en formato lquido.

*   **Regin de Mayor Venta:** La regin "TOTAL AUTOS AREA 2" tambin muestra un rendimiento destacado, casi duplicando las ventas del rea 4.

*   **Tendencia de Ventas Mensuales:** Se observa que Enero de 2022 fue el mes con las ventas ms altas en el perodo analizado. Sin embargo, la tendencia general desde principios de 2022 hasta Julio de 2023 muestra una disminucin en las ventas, lo que podra indicar desafos en el mercado o cambios en el comportamiento del consumidor.

*   **ATTR1 Dominante:** El `ATTR1` "CLORO" es claramente el que domina las ventas en ambos aos (2022 y 2023), generando ingresos considerablemente mayores en comparacin con "SAFE BLEACH", "PRELAVADOR" y "SANITIZANTE". Esto subraya la importancia estratgica de los productos con cloro en el portafolio.

*   **Marcas Lderes:** Las marcas "CLORALEX", "VANISH" y "CLOROX" se posicionan como las top 3 en trminos de ingresos totales. "CLORALEX" destaca notablemente como la marca con mayores ventas. La mayora de las marcas experimentaron una disminucin en los ingresos de 2022 a 2023, lo cual concuerda con la tendencia general de ventas observada.

Estos insights pueden ser tiles para tomar decisiones estratgicas, como enfocar esfuerzos de marketing en los formatos y regiones de mayor rendimiento, investigar las causas de la disminucin general de las ventas y capitalizar la popularidad de los productos con cloro y las marcas lderes.
